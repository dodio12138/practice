```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./words_pos.csv')
data_words = data.iloc[:, 1:]
data_words = data_words.iloc[:, :1]
data_words
```

```python
# 提取每个单词的首字母
data_words['First Letter'] = data_words['word'].str[0].str.upper()

# 提取每个单词的长度
data_words['Word Length'] = data_words['word'].apply(len)  

# 统计单词长度的词频
length_counts = data_words['Word Length'].value_counts().sort_index()

# 统计首字母的词频
letter_counts = data_words['First Letter'].value_counts().sort_index()
length_counts
```

```python
# 绘制条形图
letter_counts.plot(kind='bar', figsize=(6, 4), color='skyblue', edgecolor='black')
plt.title('Frequency of First Letters in Words')
plt.xlabel('First Letter')
plt.ylabel('Frequency')
plt.show()
```

```python
# 绘制直方图
length_counts.plot(kind='bar', figsize=(6, 4), color='skyblue', edgecolor='black')
plt.title('Distribution of Word Lengths')
plt.xlabel('Word Length')
plt.ylabel('Frequency')
plt.show()
```

```python
data_words['Letter 2'] = data_words['word'].apply(lambda x: x[1] if len(x) >= 2 else None)

letter_2_counts = data_words['Letter 2'].value_counts().sort_index()

# 绘制柱状图
plt.figure(figsize=(6, 4))
letter_2_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frequency of 2 Letters in Words')
plt.xlabel('2 Letter')
plt.ylabel('Frequency')
plt.show()
```
